```python
import os
```

```python
def count_instances(folder_path):
  """
  Counts the number of files in a folder.

  Args:
    folder_path: Path to the folder to count files.

  Returns:
    The number of files in the folder.
  """
  files = os.listdir(folder_path)
  return len(files)
```

```python
# Replace "main_directory" with the actual path to your main folder
main_directory = "foods"

# Initialize an empty dictionary to store folder-wise counts
folder_counts = {}
```

```python
for folder in os.listdir(main_directory):
  folder_path = os.path.join(main_directory, folder)
  if os.path.isdir(folder_path):
    # Call count_instances to get the number of files
    num_instances = count_instances(folder_path)
    folder_counts[folder] = num_instances
    print(f"Folder: {folder}, Instances: {num_instances}")
```

```python
# Print the final dictionary containing folder-wise instance counts
print("Folder-wise Instance Counts:")
for folder, count in folder_counts.items():
  print(f"{folder}: {count}")
```

```python
class TopKFilesDataset(torch.utils.data.Dataset):
  """
  Custom dataset class to load top K files from each folder.
  """

  def __init__(self, root_dir, k, transform=None):
    """
    Args:
      root_dir: Path to the directory containing subfolders.
      k: Number of top files to load from each subfolder.
      transform: Optional transformation to apply to loaded data (e.g., image pre-processing).
    """
    self.root_dir = root_dir
    self.k = k
    self.transform = transform

    # Initialize empty lists to store file paths and labels
    self.data = []
    self.labels = []
    self.process_folders()

  def process_folders(self):
    """
    Iterates through subfolders, collects top K file paths, and stores them.
    """
    for folder in os.listdir(self.root_dir):
      folder_path = os.path.join(self.root_dir, folder)
      if os.path.isdir(folder_path):
          # Extract label from folder name (modify logic if labels are stored elsewhere)
          label = folder  # Assuming label is the folder name

          files = os.listdir(folder_path)
          # Sort files (optional: modify for specific order)
          sorted_files = sorted(files)
          for file in sorted_files[:self.k]:
              file_path = os.path.join(folder_path, file)
              self.data.append(file_path)
              self.labels.append(label)

  def __len__(self):
    """
    Returns the total number of files loaded from all folders.
    """
    return len(self.data)

  def __getitem__(self, idx):
    """
    Loads a specific data point (file) based on the index.

    Args:
      idx: Index of the data point to load.

    Returns:
      A tuple containing the loaded data (e.g., image) and any associated label (if applicable).
    """
    image = self.data[idx]
    # image = self.transform(image)
    label = self.labels[idx]

    # return image.float(), torch.tensor(targets).float().view(1, 8)  # reshape target tensor to (1, 8)
    return image.float(), label
```

```python
k = 400  # Number of top files per folder

# Create the dataset
dataset = TopKFilesDataset(main_directory, k)
```

```python
# Access data and labels directly from the dataset object (no need to iterate)
all_data = dataset.data
all_labels = dataset.labels
```

```python
# Create a LabelEncoder object
le = LabelEncoder()
le.fit_transform(all_labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Stratified split for train-validation-test sets
X_train, X_test_val, y_train, y_test_val = train_test_split(all_data, all_labels, test_size=0.2, stratify=all_labels, random_state=42)

# Further split test-validation set while maintaining class proportions
X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5, stratify=y_test_val, random_state=42)
```
